```python
# Function to get image URLs for a model (same as before)
def get_image_urls(model_name, num_images=3):
    api_key = "AIzaSyDf0kHJe1HGJEtMMgujInAsyZhWeMVZnpI"
    cse_id = "d19b77460e3f64d31"
    url = f"https://www.googleapis.com/customsearch/v1?key={api_key}&cx={cse_id}&q={model_name}&searchType=image"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
        data = response.json()

        image_urls = []
        if 'items' in data:
            for item in data['items'][:num_images]:  # Limit to num_images
                image_urls.append(item['link'])
        return image_urls
    except requests.exceptions.RequestException as e:
        print(f"Error during API request: {e}")
        return []
    except (KeyError, TypeError) as e:
        print(f"Error parsing JSON response: {e}")
        return []

# Function to update product images
def update_product_images():
    try:
        # Find all products
        products = products_collection.find()

        for product in products:
            model_name = product['name']
            print(f"Checking images for {model_name}...")

            # Check if the product already has images
            if not product.get('images') or len(product.get('images', [])) == 0:
                print(f"Updating images for {model_name}...")

                # Get image URLs for the model
                image_urls = get_image_urls(model_name)

                # Update the product document with the image URLs
                if image_urls:
                    products_collection.update_one(
                        {"_id": product["_id"]},
                        {"$addToSet": {"images": {"$each": image_urls}}}
                    )
                    print(f"Images updated for {model_name}")
                else:
                    print(f"No images found for {model_name}")
            else:
                print(f"Skipping {model_name} as it already has images.")
    except Exception as e:
        print(f"Error updating product images: {e}")

# Call the function to update product images
update_product_images()
```

```python
def update_stock():
    try:
        # Update stock for all products
        products_collection.update_many({}, {"$set": {"stock": 25}})
        print("Stock updated for all products.")
    except Exception as e:
        print(f"Error updating stock: {e}")

# Call the function to update stock
update_stock()
```
